#### dictionray 관련 함수

```
# key list
a = {'name': 'pey', 'phone': '0119993323', 'birth': '1118'}
print('Key list')
print(a)
print(a.keys())

# value list
a = {'name': 'pey', 'phone': '0119993323', 'birth': '1118'}
print('\nValue list')
print(a)
print(a.values())

# pair list
a = {'name': 'pey', 'phone': '0119993323', 'birth': '1118'}
print('\nPair list')
print(a)
print(a.items())
```

### elif문

```
a = 33
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
```

### else 문

```
a = 200
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print("a is greater than b")
```

## Function

```
def add(a, b):
  return a + b

c = add(1, 2)
print(c)
```

```
def add(a, b):
  print(a+b)
  
add(1, 2)
```

```
def add_many(*args):
  result = 0
  for i in args:
    result += i
  return result

result = add_many(1, 2, 3)
print(result)
```

```
def add_minus(a, b):
  return a + b, a - b

c, d = add_minus(1, 2)
print(c, d)
```

## Class

- Data를 표현하는 속성(attribute)과 행위를 표현하는(method)를 포함하는 논리적인 컨테이너

```
class Calculator:
    def __init__(self):
        self.result = 0

    def add(self, num):
        self.result += num
        return self.result
```

```
cal1 = Calculator()
```

```
cal1 = Calculator()
cal2 = Calculator()

print(cal1.add(3))
print(cal1.add(4))
print(cal2.add(5))
print(cal2.add(6))
```

```
class FourCal:
  
  def setdata(self, first, second):
    self.first = first
    self.second = second
  
  def add(self):
    result = self.first + self.second
    return result

  def mul(self):
    result = self.first * self.second
    return result

  def sub(self):
    result = self.first - self.second
    return result

  def div(self):
    result = self.first / self.second
    return result
```

```
a = FourCal()
b = FourCal()
```

```
a.setdata(4, 2)
```

```
b.setdata(3, 8)
```

```
a.add()
```

```
a.mul()
```

```
a.sub()
```

```
a.div()
```
